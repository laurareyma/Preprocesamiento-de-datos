- **`codigo_postal` es `int64`**. Riesgo típico: si hay códigos postales que empiezan en 0, al convertirlos a entero se pierde el cero inicial
- **`nivel_satisfaccion` es `str`**, lo cual sugiere que no es númerico (1-5) sino texto
